```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
## create a global to store the Covid 19 Confirmed counts
#confirmed_global

confirmed_global = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'

covid19_confirmed=pd.read_csv(confirmed_global,index_col='Country/Region')

display(covid19_confirmed)
```

```python
##Transposing the content for Countries as Column; also taking only last 30 days' counts
display(covid19_confirmed.T.tail(30))
```

```python
## Finding the last date
confirmed_latest = covid19_confirmed.T.index.values[-1]
display(confirmed_latest)
```

```python
## Grouping the countries together
df_grouped_conf=covid19_confirmed.groupby('Country/Region').sum()
display(df_grouped_conf)
```

```python
## Sort the Dataframe by date and take top 10 countries for last 30 days

df_confirmed = df_grouped_conf.sort_values(confirmed_latest,ascending=False).head(10).drop(['Lat','Long'],axis=1).T.tail(30)
display(df_confirmed)
```

## Plotting this on the graph

```python
def multi_plot(df,title,addAll=True):
    fig=go.Figure()
    
    for column in df.columns.to_list():
        fig.add_trace(
            go.Scatter(
                x = df.index,
                y=df[column],
                name= column
            )
        )
    button_all = dict(label='All',
                     method='update',
                     args=[{'visible':
                            df.columns.isin(df.columns),
                            'title':'All',
                           'showlegend':True}])
    
    def create_layout_button(column):
        return dict(label = column,
                    method = 'update',
                    args=[{'visible':
                            df.columns.isin([column]),
                            'title':'column',
                           'showlegend':True}])
    
    fig.update_layout(updatemenus=[go.layout.Updatemenu(
                        active=0,
        buttons=([button_all]*addAll) + 
        list(df.columns.map(lambda column: create_layout_button(column)))
    )
    ],
                      yaxis_type="log"
    )
    
    # Update remaining layout properties
    
    fig.update_layout(title_text=title, height=800)
    
    fig.show()
```

```python
## Multiplot

multi_plot(df_confirmed, title="Logarithmic Covid19 Time Series Total Confirmed by Country")
```
